```python
import matplotlib.pyplot as plt
#%matplotlib inline # 이거는 이제 안해줘도 됨 (주피터 노트북에 내장)

plt.plot([1, 2, 3], [2, 4, 6])
plt.title("Hello plot")
plt.show()
```

### Figure와 Axes

```python
# plt.figure()는 주로 figure의 크기를 설정하는데 사용됨
plt.figure(figsize=(10, 4))
plt.plot([1, 2, 3], [2, 4, 6])
plt.title("Hello plot")
plt.show()
```

```python
figure = plt.figure(figsize=(10, 4)) # 이렇게 쓸 일은 거의 없음
print(type(figure))
```

```python
plt.figure(figsize=(10, 4), facecolor="yellow")
plt.plot([1, 2, 3], [2, 4, 6])
plt.title("Hello plot")
plt.show()
```

```python
ax = plt.axes()
print(type(ax))
```

```python
# pyplot에서 설정된 figure와 axes 객체 함께 가져오기
fig, ax = plt.subplots()
print(type(fig), type(ax))
```

### 여러 개의 plot을 가지는 figure 설정

```python
import numpy as np

x_value = [1, 2, 3, 4]
y_value = [2, 4, 6, 8]
x_value = np.array([1, 2, 3, 4])
y_value = np.array([2, 4, 6, 8])

# 입력값으로 파이썬 리스트, ndarray 모두 가능. 단, x축/y축 개수가 같아야 함
plt.plot(x_value, y_value)
```

```python
import pandas as pd

df = pd.DataFrame({'x_value': [1, 2, 3, 4],
                  'y_value': [2, 4, 6, 8]})

# 입력값으로 pandas series 및 dataframe도 가능
plt.plot(df['x_value'], df['y_value'])
```

```python
plt.plot(x_value, y_value, color="green")
```

```python
# 그 외 다양한 인자들이 있음
plt.plot(x_value, y_value, color='red', marker='o', linestyle='dashed', linewidth=2, markersize=12)
```

### x축, y축에 축명을 텍스트로 할당 (xlabel, ylabel)

```python
plt.plot(x_value, y_value, marker='o', linestyle="dotted", color="green")
plt.xlabel("x_val")
plt.ylabel("y_val")
plt.show()
```

### x축, y축 틱값을 회전시켜서 보여줌
### x축 값이 문자열이고 많은 tick값이 있을 때 적용

```python
x_value = np.arange(1, 100)
y_value = x_value * 2

plt.plot(x_value, y_value, linestyle="dashed")
plt.xticks(rotation=45)
plt.title("Hello plot")

plt.show()
```

```python
plt.plot(x_value, y_value, color="green")

plt.xlabel("x axis")
plt.ylabel("y axis")

plt.xticks(ticks=np.arange(0, 100, 5), rotation=90)
plt.yticks(rotation=45)

plt.title("Hello plot")

plt.show()
```

### xlim(), ylim()

```python
plt.plot(x_value, y_value, color="green")

plt.xlabel("x axis")
plt.ylabel("y axis")

# x축 값을 0~50으로, y축 값을 0~100으로 제한
plt.xlim(0, 50)
plt.ylim(0, 100)

plt.title("Hello plot")

plt.show()
```

### 범례(legend) 설정하기

```python
plt.plot(x_value, y_value, color="green", label="temp")

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Hello plot")
plt.legend()

plt.show()
```

### matplotlib으로 여러 개의 plot을 하나의 axes내에서 그릴 수 있음

```python
x_value_01 = np.arange(1, 100)
y_value_01 = x_value_01 * 2
y_value_02 = x_value_01 * 4

plt.plot(x_value_01, y_value_01, color="green", label="temp1")
plt.plot(x_value_01, y_value_02, color="red", label="temp2")

plt.title("Hello plot")
plt.legend()

plt.show()
```

```python
x_value_01 = np.arange(1, 10)
y_value_01 = x_value_01 * 2
y_value_02 = x_value_01 * 4

plt.plot(x_value_01, y_value_01, color="red", marker="o", linestyle="--", label="red")
plt.bar(x_value_01, y_value_02, color="green", label="green")

plt.title("Hello Plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()

plt.show()
# plt.bar(x_value_01, y_value_02, color="red", label="temp2")
```

### Axes 객체에서 직접 작업

```python
figure = plt.figure(figsize=(10, 6))
ax = plt.axes()

ax.plot(x_value_01, y_value_01, color="green", linestyle="dotted", marker="o", label="green")
ax.bar(x_value_01, y_value_01, color="yellow", label="yello")

ax.legend() # set_legend()가 아닌 그냥 legend()
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_title("Hello plot")

plt.show()
```

### 여러 개의 subplots을 가지는 시각화

```python
x_value_01 = np.arange(1, 10)
x_value_02 = np.arange(1, 20)
y_value_01 = x_value_01 * 2
y_value_02 = x_value_02 * 2

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

ax1.plot(x_value_01, y_value_01, color="red", linestyle="dotted", marker="o", label="temp1")
ax2.bar(x_value_02, y_value_02, color="green", label="temp2")

ax1.set_title("Hello plot1")
ax2.set_title("Hello plot2")

ax1.set_xlabel("X axis")
ax1.set_ylabel("Y axis")
ax2.set_xlabel("X axis")
ax2.set_ylabel("Y axis")

ax1.legend()
ax2.legend()

plt.show()
```

```python
import numpy as np

x_value_01 = np.arange(1, 10)
x_value_02 = np.arange(1, 20)
y_value_01 = 2 * x_value_01
y_value_02 = 2 * x_value_02

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

ax[0].plot(x_value_01, y_value_01, color="red", linestyle="dotted", marker="o", label="temp1")
ax[1].bar(x_value_02, y_value_02, color="green", label="temp2")

ax[0].set_title("Hello plot1")
ax[1].set_title("Hello plot2")

ax[0].set_xlabel("X axis")
ax[0].set_ylabel("Y axis")
ax[1].set_xlabel("X axis")
ax[1].set_ylabel("Y axis")

ax[0].legend()
ax[1].legend()

plt.show()
```

```python
import numpy as np

x_value_01 = np.arange(1, 10)
x_value_02 = np.arange(1, 20)
y_value_01 = 2 * x_value_01
y_value_02 = 2 * x_value_02

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(12, 6))

ax[0][0].plot(x_value_01, y_value_01, color='red', marker='o', linestyle='dashed', linewidth=2, markersize=6, label='temp_01')
ax[0][1].bar(x_value_02, y_value_02, color='green', label='temp_02')
ax[1][0].plot(x_value_01, y_value_01, color='green', marker='o', linestyle='dashed', linewidth=2, markersize=6, label='temp_03')
ax[1][1].bar(x_value_02, y_value_02, color='red', label='temp_04')

ax[0][0].set_xlabel('ax[0][0] x axis')
ax[0][1].set_xlabel('ax[0][1] x axis')
ax[1][0].set_xlabel('ax[1][0] x axis')
ax[1][1].set_xlabel('ax[1][1] x axis')

ax[0][0].legend()
ax[0][1].legend() 
ax[1][0].legend()
ax[1][1].legend() 

plt.show()
```
